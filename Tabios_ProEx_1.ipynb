```python
import requests as rq 
from io import StringIO
import pandas as pd

url = 'https://raw.githubusercontent.com/ABRION77/DS-100-repository/main/WHR_2019.csv'
response = rq.get(url)
data = pd.read_csv(StringIO(response.text))
pd.set_option('display.max_rows', None)
data.head()
```

```python
# Filtering based on GDP per capita
happiness_data_filtered = data[data['GDP per capita'].between(0.8, 1.5)].copy()

# Calculating Generosity
happiness_data_filtered['Generosity'] = np.round(happiness_data_filtered['GDP per capita'] / happiness_data_filtered['Social support'], 3)
happiness_data_filtered['Generosity'] = happiness_data_filtered['Generosity'].fillna(0)

# Selecting required coluimns
happiness_data_filtered = happiness_data_filtered[['Overall rank', 'Country or region', 'GDP per capita', 'Healthy life expectancy', 'Generosity', 'Perceptions of corruption']]

# Display the first 25 rows for checking
print(happiness_data_filtered.head(24))
```

```python
# Calculation side of the programming exercises one statistical and the other for standard deviation
import numpy as np

# Filtering based on GDP per capita
happiness_data_filtered = data[data['GDP per capita'].between(0.8, 1.5)].copy()

# Equation 1: Calculating Generosity
happiness_data_filtered['Generosity'] = np.round(happiness_data_filtered['GDP per capita'] / happiness_data_filtered['Social support'], 3).fillna(0)

# Selecting Required Columns
happiness_data_filtered = happiness_data_filtered[['Overall rank', 'Country or region', 'GDP per capita', 'Healthy life expectancy', 'Generosity', 'Perceptions of corruption']]

# For checking display first few rows
print(happiness_data_filtered.head(24))

# Calculate statistics
stats_columns = ['Generosity', 'Perceptions of corruption', 'Healthy life expectancy']
statistics = {}

for column in stats_columns:
    mean = happiness_data_filtered[column].mean()
    median = happiness_data_filtered[column].median()
    mode = happiness_data_filtered[column].mode()[0]
    std = happiness_data_filtered[column].std()
    average = (mean + median + mode) / 3
    statistics[column] = {'Mean': mean, 'Median': median, 'Mode': mode, 'Average': average, 'Standard Deviation': std}

# Print/Display Statistics
for column, values in statistics.items():
    print(f"{column} statistics:")
    for stat, value in values.items():
        print(f"{stat}: {value}")
    print()
```
